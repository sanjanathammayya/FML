```python
#Fitting Simple LR to the Training set
from sklearn.linear_model import LinearRegression # when we give fit it wiull train the model
regressor=LinearRegression() # regressor object for LR
regressor.fit(X_train,y_train)
```

```python
# predicting the test set results
y_pred=regressor.predict(X_test) # while training give only X_test not y_test
```

```python
# Visualising the training set results
plt.scatter(X_train,y_train,color='green') # train data in green line # y_train is y hat
plt.scatter(X_test,y_test,color='red') # test data in red line
plt.plot(X_train,regressor.predict(X_train),color='blue') # gives o/p of y hat. from  traning I have to predict the value ie LR line
plt.title ('Salary vs Experience (Training set)')
plt.xlabel ('year of Experience')
plt.ylabel ('Salary')

plt.show()
```
